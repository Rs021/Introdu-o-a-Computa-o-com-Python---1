```python
import random

dicta = '-+_@#$%[]{}ABCDEFGHIJKLMNOPQRSTUVWYXZ-+_@#$%[]{}abcdefghijklmnopqrstuvwyxz123456789-+_@#$%[]{}'





def debug(user,passw):
    file = open('debug.txt', 'w')
    file.write(f"User: {user} | Pass: {passw}")
   

def genpass(dicta):
    password = ''
    while len(password) < 12:
        for i in range(random.randrange(0,len(dicta))):
           password+= dicta[random.randint(0,i)]
    return password

dbg = ''
for i in range(10):
    
    dbg += genpass(dicta) + '\n'

for pos,char in enumerate(dbg):
    print(char + '\n')
```
